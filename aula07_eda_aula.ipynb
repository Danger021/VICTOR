## 1 - Carregar os arquivos de cadastro, investimento1 e investimento2, e mostrar a quantidade de linhas de cada um, além das informações desses dataframes (por exemplo, o tipo dos dados das coluna)

```python
import pandas as pd
import datetime as dt
import numpy as np
pd.options.display.max_columns = 35 #para aumentar a quantidade de colunas a serem visualizadas

urls = ['https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/cadastro.csv',
'https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/investimento_parte1.csv',
 'https://raw.githubusercontent.com/renatapink/DSdegree/master/Modulo2/Mini_projeto1/investimento_parte2.csv']
cadastro_df = pd.read_csv(urls[0])
investimento_parte1_df = pd.read_csv(urls[1])
investimento_parte2_df = pd.read_csv(urls[2])

print("Quantidade de linhas e informações dos DataFrames:")
print(f"\nCadastro DataFrame: {cadastro_df.shape[0]} linhas")
print(cadastro_df.info())

print(f"\nInvestimento Parte 1 DataFrame: {investimento_parte1_df.shape[0]} linhas")
print(investimento_parte1_df.info())

print(f"\nInvestimento Parte 2 DataFrame: {investimento_parte2_df.shape[0]} linhas")
print(investimento_parte2_df.info())
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

Criar um novo dataframe  
Dica: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
df_juntos = pd.concat([investimento_parte1_df, investimento_parte2_df], ignore_index = True)
df_juntos.shape[0]
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas do item anterior

```python
df_unico_juntos = df_juntos.drop_duplicates()
df_unico_juntos.shape[0]
```

## 4 - Junte o dataframe de investimento com o de cadastro, criando um novo dataframe
Dica: use a coluna id como chave para juntar os dois dataframes

```python
df_completos = pd.merge(df_unico_juntos,cadastro_df)
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
df_completos.sort_values ('id',ascending = True)
df_completos.reset_index(drop = True, inplace = True)
df_completos
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
missing_data = df_completos.isnull().sum()
missing_percent = (df_completos.isnull().sum() / len(df_completos))*100

# Criar um DataFrame para visualizar melhor os dados faltantes e suas porcentagens
missing_df = pd.DataFrame({'Total Missing': missing_data, 'Percentage Missing': missing_percent})
missing_df = missing_df[missing_df['Total Missing'] > 0]  # Filtrar apenas colunas com dados faltantes
print(missing_df)

# Preenchendo valores faltantes para Ativo01 a Ativo05 com a mediana
ativos = ['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06']
for ativo in ativos:
    if ativo in df_completos.columns:
        df_completos[ativo].fillna(df_completos[ativo].median(), inplace=True)

# Preenchendo Nota1 e Nota2 com a média
notas = ['Nota1', 'Nota2']
for nota in notas:
    if nota in df_completos.columns:
        df_completos[nota].fillna(df_completos[nota].mean(), inplace=True)
print(df_completos)
```

## 8 - Calcule os seguintes dados
- Calcule o total investido em cada coluna Ativo (tente usar o apply)
- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo
- Qual o nome e a idade do investidor mais velho?
- Qual a mediana da coluna Média?

```python
ativos_columns = df_completos.filter(like='Ativo').columns
total_investido = df_completos[ativos_columns].apply(np.sum, axis=0)
print(total_investido)

# Calcular o menor valor investido em cada ativo
menor_investido = df_completos[ativos_columns].apply(np.min, axis=0)
print(menor_investido)
```

```python
# Calcular a média e o desvio padrão de cada ativo
media_ativos = df_completos[ativos_columns].apply(np.mean, axis=0)
desvio_padrao_ativos = df_completos[ativos_columns].apply(np.std, axis=0)

print("Médias:\n", media_ativos)
print("Desvio Padrão:\n", desvio_padrao_ativos)
```

```python
# Encontrar o investidor mais velho
investidor_mais_velho = df_completos.loc[df_completos['Idade'].idxmax(), ['Nome', 'Idade']]
print(investidor_mais_velho)
```

```python
# Calcular a mediana da coluna 'Média'
mediana_media = media_ativos.median()
print("Mediana da coluna Média:", mediana_media)
```

## Bonus - Leitura de Arquivos Grandes

```python
# Contar quantos investidores existem em cada Perfil
investidores_perfil = df_completos['Perfil'].value_counts()
print(investidores_perfil)

# Filtrar investidores com Perfil Agressivo e Moderado
agressivos_moderados = df_completos[df_completos['Perfil'].isin(['Agressivo', 'Moderado'])]

# Mostrar os 10 primeiros IDs
primeiros_ids = agressivos_moderados['id'].head(10)
print(primeiros_ids)

# Contar quantos investidores são do sexo feminino
investidores_feminino = df_completos[df_completos['Sexo'] == 'Feminino'].shape[0]
print(investidores_feminino)

# Filtrar investidores que investiram mais de 16000 no Ativo04
investidores_ativo04 = df_completos[df_completos['Ativo04'] > 16000].shape[0]
print(investidores_ativo04)

# Filtrar investidores com os critérios de Ativo04 e Ativo05
investidores_ativos_04_05 = df_completos[(df_completos['Ativo04'] > 16000) & (df_completos['Ativo05'] < 5000)].shape[0]
print(investidores_ativos_04_05)

# Criar uma coluna de investimento total
df_completos['total_investido'] = df_completos[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].sum(axis=1)

# Filtrar as investidoras femininas com mais de 25 anos e investimento menor que 10000
investidoras_femininas = df_completos[(df_completos['Sexo'] == 'Feminino') & 
                                      (df_completos['Idade'] > 25) & 
                                      (df_completos['total_investido'] < 10000)]
print(investidoras_femininas[['Nome', 'Idade', 'total_investido']])

# Contar quantos investidores existem em cada esIdadado
investidores_estado = df_completos.groupby('Estado')['id'].count()
print(investidores_estado)

# Somar o total investido por estado
soma_investimento_estado = df_completos.groupby('Estado')['total_investido'].sum()
print(soma_investimento_estado)

# Calcular a média do total investido por tipo de cliente
media_investido_tipo = df_completos.groupby('Perfil')['total_investido'].mean()
print(media_investido_tipo)

# Divisão por estado
investimento_tipo_estado = df_completos.groupby(['Estado', 'Perfil'])['total_investido'].sum()
print(investimento_tipo_estado)

#
soma_investido_por_estado = df_completos.groupby('Estado')['total_investido'].sum()
print(soma_investido_por_estado)

# Calculando a média do total investido por categoria
media_investido_por_categoria = df_completos.groupby('categoria')['total_investido'].mean()
print("Média do total investido por categoria:")
print(media_investido_por_categoria)
```

```python
dummies_perfil = pd.get_dummies(df_completos['Perfil'], prefix='Perfil')
df_completos = pd.concat([df_completos, dummies_perfil], axis=1)

print("DataFrame com colunas dummies:")
print(df_completos.head())
```

```python
# Criação da tabela com total de investimentos por perfil e categoria
tabela_investimentos = pd.pivot_table(
    df_completos,
    values='total_investido',
    index=dummies_perfil.columns,
    columns='categoria',
    aggfunc='sum',
    fill_value=0
)

print("Tabela com total de investimentos por perfil e categoria:")
print(tabela_investimentos)
```
